# **The Iris dataset contains a set of 150 records under 4 Features: (Petal Length, Petal Width, Sepal Length, Sepal width) and Target: 3 Class(Species): (Iris Setosa, Iris Versicolour, Iris Virginica)**

# **Import Libraries:**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.metrics import accuracy_score, confusion_matrix
```

# **Import & show Data**

```python
iris = load_iris()
df = pd.DataFrame(iris.data, columns = iris.feature_names)
df.head(10)
```

```python
iris.feature_names
```

```python
iris.target_names
```

```python
df.hist(figsize = (12,10))
plt.show()
```

# **Prepare Data**

```python
X = iris.data
print(np.shape(X))
X
```

```python
y = iris.target
print(np.shape(y))
y
```

```python
X_train, X_test, y_train, y_test= train_test_split(X, y, test_size=0.3, random_state =32)
```

# **Decision Tree Classifier Model:**

```python
DTC = DecisionTreeClassifier(max_depth=5)
DTC.fit(X_train,y_train)
```

```python
y_pred=DTC.predict(X_test)
y_pred
```

```python
y_test
```

```python
df_new = pd.DataFrame({'Actual': y_test, 'Predicted': y_pred})
df_new
```

**Predict New value**

```python
X_val = [6.1, 2.7, 3.9, 1.2]
y_pred_new = DTC.predict([X_val])
y_pred_new
```

**Evaluat Model**

```python
accuracy = accuracy_score(y_test, y_pred)
print('The accuracy Score of Tree:', accuracy)
```

```python
confusion_matrix(y_test, y_pred)
```

# **Visualizing the tree**

```python
plt.figure(figsize=(15,12))
plot_tree(DTC, filled=True)
plt.show()
```

```python
plt.figure(figsize=(17,12))
plot_tree(DTC, filled=True, 
          feature_names=["Sepal Length", "Sepal Width", "Petal Length", "Petal Width"], 
          class_names=["Setosa", "Versicolour", "Virginica"])
plt.show()
```
